```python
import pandas as pd
import numpy as np
import joblib
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"E:\salary prediction\data\salary_prediction_data.csv")
df
```

```python
df
x = df.drop('final_salary', axis = 1)
y = df['final_salary']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=10)
```

```python
lin_reg_model = LinearRegression()
lin_reg_model.fit(x_train, y_train)
```

```python
lin_reg_model.coef_
```

```python
lin_reg_model.intercept_
```

```python
y_pred_test = lin_reg_model.predict(x_test)
```

```python
y_test
```

```python
y_pred_test[:5]
```

```python
y_pred_test = lin_reg_model.predict(x_test)
mse = mean_squared_error(y_test, y_pred_test)
print("MSE :",mse)
rmse = np.sqrt(mse)
print("RMSE :", rmse)
mae = mean_absolute_error(y_test, y_pred_test)
print("MAE :",mae)
r2_value = r2_score(y_test, y_pred_test)
print("R-squared :",r2_value)
```

```python
y_pred_train = lin_reg_model.predict(x_train)
mse = mean_squared_error(y_train, y_pred_train)
print("MSE :",mse)
rmse = np.sqrt(mse)
print("RMSE :", rmse)
mae = mean_absolute_error(y_train, y_pred_train)
print("MAE :",mae)
r2_value = r2_score(y_train, y_pred_train)
print("R-squared :",r2_value)
```

```python
BASE_DIR = os.path.dirname(os.path.abspath(r"E:\salary prediction\data"))
MODEL_PATH = os.path.join(BASE_DIR, "salary_model.pkl")
joblib.dump(lin_reg_model, MODEL_PATH)
```
